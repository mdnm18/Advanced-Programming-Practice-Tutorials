# Symbolic programming paradigm
Symbolic Programming is a programming paradigm where the program manipulates symbols and expressions directly, rather than just numbers or text.

We study the Symbolic Programming Paradigm to understand how computers can manipulate symbols, reason logically, and solve abstract problems — forming the foundation of AI, mathematics, and reasoning systems.





In Symbolic Programming, programs manipulate symbols and expressions, not just numbers.
It’s used in:

1. AI (reasoning and logic)

2. Mathematics (algebra, calculus)

3. Compilers (expression trees)

4. Expert systems

```python
# symbolic program for AI reasoning

def fever_and_cough(has_fever, has_cough):
    if has_fever and has_cough:
        return "Possible Flu"
    elif has_fever:
        return "Possible Infection"
    elif has_cough:
        return "Possible Cold"
    else:
        return "Healthy"

print(fever_and_cough(True, True))
print(fever_and_cough(True, False))
print(fever_and_cough(False, True))
```

```python
# Symbolic program for mathematical expressions
import sympy as sp

# Define symbols
x, y = sp.symbols('x y')

# Create a symbolic expression
expr = (x + y)**2

# Expand it
expanded = sp.expand(expr)
print("Expanded:", expanded)

# Differentiate with respect to x
diff_expr = sp.diff(expr, x)
print("Derivative w.r.t x:", diff_expr)

# Substitute a value
value = expr.subs({x: 2, y: 3})
print("After substitution:", value)
```

```python
#Develop a Python program using the SymPy library to compute both the derivative and integral of the given mathematical expression:

from sympy import symbols, diff, integrate

# Define the variable
x = symbols('x')

# Define the expression
f = 3*x**3 - 5*x**2 + 2*x - 7

# Compute derivative
derivative_f = diff(f, x)

# Compute integral
integral_f = integrate(f, x)

# Display results
print("Given Expression: ", f)
print("Derivative of f(x): ", derivative_f)
print("Integral of f(x): ", integral_f)
```

```python
# Develop a Python program using the SymPy library to compute the derivative and integral of the function


# Import required functions from SymPy
from sympy import symbols, diff, integrate

# Define the variable
x = symbols('x')

# Define the function f(x)
f = 2*x**4 - 3*x**2 + 5

# Compute the derivative of f(x)
f_derivative = diff(f, x)

# Compute the indefinite integral of f(x)
f_integral = integrate(f, x)

# Display results
print("Given function f(x) =", f)
print("Derivative f'(x) =", f_derivative)
print("Integral ∫f(x) dx =", f_integral)
```

```python

# Using SymPy, find the derivative of f(x)=sin(x)⋅cos(x) and the definite integral of f(x)from 0 to π/2.

# Import necessary functions from SymPy
from sympy import symbols, sin, cos, diff, integrate, pi

# Define the variable
x = symbols('x')

# Define the function f(x)
f = sin(x) * cos(x)

# Compute the derivative of f(x)
f_derivative = diff(f, x)

# Compute the definite integral of f(x) from 0 to π/2
f_integral = integrate(f, (x, 0, pi/2))

# Display results
print("Given function f(x) =", f)
print("Derivative f'(x) =", f_derivative)
print("Definite integral of f(x) from 0 to π/2 =", f_integral)
```
