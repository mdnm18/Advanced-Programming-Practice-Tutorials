# Automata Based Programming Paradigm
Automata-based programming is a programming paradigm where the logic of the program is designed as a finite state machine (FSM) — meaning the program moves between a finite number of states based on inputs or events.
# In this paradigm:

1. A program is divided into states (specific situations or modes of operation).

2. It receives inputs/events (like button clicks, commands, or signals).

3. Based on input, it transitions from one state to another.

4. The program’s behavior depends on its current state.

# Simple Example – Door System

Imagine an automatic door:

States: CLOSED, OPEN

Input: sensor detects person, no person detected

image.png



```python
class Door:
    def __init__(self):
        self.state = "CLOSED"

    def on_event(self, event):
        if self.state == "CLOSED" and event == "person_detected":
            self.state = "OPEN"
        elif self.state == "OPEN" and event == "no_person":
            self.state = "CLOSED"

    def __str__(self):
        return f"Door is currently {self.state}"

# Test
door = Door()
print(door)
door.on_event("person_detected")
print(door)
door.on_event("no_person")
print(door)
```

# Here the Door behaves like a Finite State Machine (FSM).

Finite Automata (FA)

A Finite Automaton (FA) is like a small machine that reads an input string (symbols like 0,1 or a,b) one by one and decides whether to accept or reject the string.

It has:

1. Finite number of states (like checkpoints)

2. A start state

3. One or more accepting (final) states

Transitions (rules for moving between states based on the input symbol)

Example: Robot that reads each symbol and moves between rooms (states) based on what it reads.
# Deterministic Finite Automata (DFA)
A DFA is a type of finite automaton where —
for every state and every input symbol, there is exactly one possible next state.
That means:
➡ No confusion, no choices.
➡ One state → one input → one next state.
# Non-Deterministic Finite Automata (NFA)
An NFA is a type of FA where —
1. For some state and input, there may be multiple possible next states (or even none).
2. It can also move to a new state without consuming any input symbol (ε-transition).

So the machine can “guess” or “branch” into several possibilities.

```python
# Write a Python program using the Automata library to simulate a DFA that accepts binary strings ending with 11.
# • Define states, input symbols, transitions, initial state, and final states.
# • Test the DFA with the inputs: '11', '1011', '1100'.

from automata.fa.dfa import DFA

# Define the DFA that accepts strings ending with '11'
dfa = DFA(
    states={'q0', 'q1', 'q2'},
    input_symbols={'0', '1'},
    transitions={
        'q0': {'0': 'q0', '1': 'q1'},
        'q1': {'0': 'q0', '1': 'q2'},
        'q2': {'0': 'q0', '1': 'q2'}
    },
    initial_state='q0',
    final_states={'q2'}
)

# Test input strings
test_strings = ['11', '1011', '1100']

# Evaluate and print results
for s in test_strings:
    if dfa.accepts_input(s):
        print(f"The string '{s}' is ACCEPTED by the DFA.")
    else:
        print(f"The string '{s}' is REJECTED by the DFA.")
```

```python
# Construct an NFA over the alphabet {0,1} that accepts all binary strings ending with 010.
# • Define the NFA formally and implement it in Python.
# • Test the NFA on the inputs: '11010', '1010', '1001'.
from automata.fa.nfa import NFA

# Define the NFA that accepts strings ending with '010'
nfa = NFA(
    states={'q0', 'q1', 'q2', 'q3'},
    input_symbols={'0', '1'},
    transitions={
        'q0': {'0': {'q0', 'q1'}, '1': {'q0'}},
        'q1': {'1': {'q2'}},
        'q2': {'0': {'q3'}},
        'q3': {'0': {'q0', 'q1'}, '1': {'q0'}}
    },
    initial_state='q0',
    final_states={'q3'}
)

# Test strings
test_strings = ['11010', '1010', '1001']

# Check and print results
for s in test_strings:
    if nfa.accepts_input(s):
        print(f"The string '{s}' is ACCEPTED by the NFA.")
    else:
        print(f"The string '{s}' is REJECTED by the NFA.")

```
